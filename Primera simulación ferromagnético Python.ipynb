```python
import numpy as np
import matplotlib.pyplot as mt
```

```python
#PROGRAMA--------------------------------------------------

N=1000 #Número par
s=1/2#Spin individual
S=int(N*s)#Spin total
Ms=np.arange(S, -S - 1, -1)

#Matriz S_z----------------
S_z=np.diag(Ms)

#Matriz S_+----------------


def s_mas(m):
    return np.sqrt(S*(S+1)-m*(m+1))

S_mas = np.zeros((2*S+1, 2*S+1))
s_mas_vector= s_mas(np.arange(S-1, -S - 1, -1))

for i in range(len(s_mas_vector)):
    S_mas[i, i + 1] = s_mas_vector[i]

#Matriz S_- -------------    

def s_menos(m):
    return np.sqrt(S*(S+1)-m*(m-1))

S_menos = np.zeros((2*S+1, 2*S+1))
s_menos_vector= s_menos(np.arange(S, -S, -1))
for i in range(len(s_menos_vector)):
    S_menos[i+1, i] = s_menos_vector[i]


#Matriz S_x

S_x=1/2 * (S_mas + S_menos)
```

```python
#Hamiltoniano considerando únicamente el término cuadrático de S_z y el acoplo con el campo magnético h
J=1
h=0.1
H=-(4*J)/N * np.diag(Ms*Ms)-2*h*S_x

#Obtenemos las autoenergías diagonalizando la matriz utilizando la función linalg.eig de numpy:
def diagonalizar_matriz(matriz):
    # Calcular autovalores y autovectores
    autovalores, autovectores = np.linalg.eig(matriz)
    # Construir la matriz diagonal con los autovalores
    matriz_diagonal = np.diag(autovalores)
    
    return autovalores, matriz_diagonal, autovectores

autoenergias=diagonalizar_matriz(H)[0]

print(f'Los autovalores del hamiltoniano son: {sorted(autoenergias)}')


```

```python
#Hamiltoniano de una cadena cuántica ferromagnética de largo alcance de espín-1/2 con interacciones de 4 espines

J=1
h=1
K=0.05

H=-(4*J)/N * np.diag(Ms*Ms)-2*h*S_x-(16*K)/(N**3)*np.diag(Ms*Ms*Ms*Ms)

def diagonalizar_matriz(matriz):
    # Calcular autovalores y autovectores
    autovalores, autovectores = np.linalg.eig(matriz)
    # Construir la matriz diagonal con los autovalores
    matriz_diagonal = np.diag(autovalores)
    
    return autovalores, matriz_diagonal, autovectores

#Obtenemos autoenergías
autovalores=diagonalizar_matriz(H)[0]
autovectores=diagonalizar_matriz(H)[2]
print(sorted(autovalores))


```

```python
#Dada el alto tiempo de cálculo incluso para órdenes de magnitud de N de 10^3, buscamos mejorar este tiempo utilizando propiedades
#de la matriz que queremos diagonalizar, en este caso la tridiagonalidad.

J=1
h=1
K=0.05

H=-(4*J)/N * np.diag(Ms*Ms)-2*h*S_x-(16*K)/(N**3)*np.diag(Ms*Ms*Ms*Ms)


from scipy.linalg import eigh_tridiagonal
def diagonalizar_tridiagonal_simetrica(diagonal_principal, diagonal_superior):
    # Calcular autovalores y autovectores
    autovalores, autovectores = eigh_tridiagonal(diagonal_principal, diagonal_superior, eigvals_only=False)
    return autovalores,autovectores

#Utilización
diagonal_principal = np.diag(H)
diagonal_superior = np.diag(H, k=1)


autovalores, autovectores = diagonalizar_tridiagonal_simetrica(diagonal_principal, diagonal_superior)
print(sorted(autovalores))



```
